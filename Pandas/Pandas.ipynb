## Visualization In Pandas

```python
#series and dataframe line chart
import numpy as np
import matplotlib.pyplot as plt
ts= pd.Series(np.random.randn(50),index=pd.date_range('today',periods=50))
ts=ts.cumsum()
ts.plot()
plt.xlabel("Date")
plt.ylabel("Value")
plt.grid()
plt.show()
```

```python
df=pd.DataFrame(np.random.randn(50,4),index=ts.index,columns=['a','b','c','d'])
df=df.cumsum()
df.plot()
```

## Remove Repeated Data using Pandas

```python
import pandas as pd
df7 = pd.DataFrame({'K': [1, 2, 2, 2, 3, 4, 3, 5, 6, 7, 4, 6, 8, 9, 6, 4, 6]})
df7_sorted = df7.sort_values(by='K').reset_index(drop=True)
# Filter rows where consecutive values in 'K' are different
unique_values_df = df7_sorted.loc[df7_sorted['K'].shift() != df7_sorted['K']]
print(unique_values_df)
```
