# Matplotlib
### What is Matplotlib?
- It is an opensouce drawing library which supports rich drawing types
- it is used to draw 2D and 3D Graphics


```python
from matplotlib import pylab
```

```python
import numpy as np
x=np.linspace(0,10,30)
y=x*x+3
print(x)
print(y)
```

```python
pylab.plot(x,y,'r') #'r' stands for red, red is generally used for loss
```

```python
pylab.subplot(1,3,1) #the contents of the brackets represents(rows,columns,indexes)
pylab.plot(x,y,"r--") #the third parameter here determines colour and line style
pylab.subplot(1,3,2)
pylab.plot(y,x,'g+-')
pylab.subplot(1,3,3)
pylab.plot(x+2,y-1,"y-+")
```

```python
from matplotlib import pyplot as plt
```

```python
fig= plt.figure()
axis = fig.add_axes([0.1,0.5,0.5,0.5]) #cotrol the left, right, width, height of the canvas (from 0 to 1)
axis.plot(x,y,'r')
```

#### Note- If the value of row or column in below is 1 then it is a 1D array hence we can use plot func. but in 2D we have to use 2 for loops

```python
# again we can draw subgraphs 
fig , axes = plt.subplots(nrows = 2, ncols =1) #submap is of 1 row , 2 column
for i in axes:
    i.plot(x,y,'r')
```

```python
  # again we can draw subgraphs
fig , axes = plt.subplots(nrows = 2, ncols =3) #submap is of 1 row , 2 column
for row in axes:
    for i in row:
        i.plot(x,y,'r')
```

```python
#for insertion of a graph in another. 
fig2 = plt.figure()
axes1 = fig2.add_axes([0.1,0.1,0.8,0.8])# big canvas
axes2 = fig2.add_axes([0.2,0.5,0.4,0.3])# small Canvas

axes1.plot(x,y,'r')
axes2.plot(y,x,'g')
```

```python
fig, axes = plt.subplots(figsize=(4,4))
axes.set_title('Title of the Graph')
axes.set_xlabel('Age')
axes.set_ylabel('time')
axes.plot(x,x**2, 'r')
axes.plot(x,x**3, 'g')
axes.legend(['y=x**2','y=x**3'],loc=2)
```

```python
fig5, ax = plt.subplots(figsize=(4,4), dpi=100)

# Plot multiple lines with different styles on the same axes
ax.plot(x, x + 1, color='teal', lw=2, linestyle='-')    
ax.plot(x, x + 2, color='yellow', lw=3, linestyle='-.')  
ax.plot(x, x + 3, color='red', lw=1.5, linestyle=':') 
ax.plot(x, x + 4,color = 'blue',marker='o')
ax.plot(x, x + 5,color = 'deeppink',marker='s')
ax.plot(x, x + 6,color = 'black',marker='s',markerfacecolor='red',markersize=2)
ax.plot(x, x + 7,color = 'green',marker='s',markerfacecolor='red',)
```

### Other 2D graphs

```python
n=np.array([0,1,2,3,4,5])
fig, Axes = plt.subplots(1,4, figsize=(16,5))
Axes[0].set_title('scatter')
Axes[0].scatter(x,x+0.25*np.random.randn(len(x)))
Axes[1].set_title('STEP')
Axes[1].step(n,n**2,lw=2)
Axes[2].set_title('Bar Plot')
Axes[2].bar(n,n**2,align='center',alpha=0.8)
Axes[3].set_title('fill_between')
Axes[3].fill_between(x,x**2,x**3,color='green')
```

```python
# Draw a Histogram
n= np.random.randn(100000)
fig , axes = plt.subplots(1,3,figsize=(12,4))
axes[0].set_title("Default Histogram")
axes[0].hist(n)
axes[1].set_title("Cumulative Detailed Histogram")
axes[1].hist(n, cumulative=True, bins=50)
axes[2].set_title("Cumulative Detailed Histogram")
axes[2].hist(n, cumulative=True, bins=8)
```

### Draw contour image

```python
import matplotlib.cm as cm #colour map
import matplotlib.pyplot as plt
delta=0.025
x=np.arange(-3.0,3.0,delta)
y=np.arange(-2.0,2.0,delta)
A,B=np.meshgrid(x,y)
z1=np.exp(-A**2 - B**2)
z2=np.exp(-(A-1)**2 - (B-1)**2)
z=(z1-z2)*2
print(A)
print(B)
```

```python
fig, ax =plt.subplots()
cs=ax.contour(x,y,z)
ax.clabel(cs,inline=1,fontsize=10)
ax.set_title("contour map")
```

### Draw 3D surface image

```python
from mpl_toolkits.mplot3d.axes3d import Axes3D
fig = plt.figure(figsize=(14,6))
#specify the 3D graphics to draw with projection = '3d'
ax=fig.add_subplot(1,2,1,projection='3d')
ax.plot_surface(A,B,z,rstride=4,cstride=4,linewidth=0)
```
